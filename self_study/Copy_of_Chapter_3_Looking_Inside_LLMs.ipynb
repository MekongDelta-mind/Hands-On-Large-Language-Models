# Loading the LLM

````python
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline
````

````python
# Load model and tokenizer
tokenizer = AutoTokenizer.from_pretrained("microsoft/Phi-3-mini-4k-instruct")
````

````python
model = AutoModelForCausalLM.from_pretrained(
    "microsoft/Phi-3-mini-4k-instruct",
    device_map="cuda",
    torch_dtype="auto",
    trust_remote_code=True,
)
````

HOw to know when to use the trust_remote_code or not ?

````python
# Create a pipeline
generator = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    return_full_text=False,
    max_new_tokens=50,
    do_sample=False,
)
````

# The Inputs and Outputs of a Trained Transformer LLM

### 1. **Direct Inference (End-to-End Model)**:
Normally, when you call a language model end-to-end, like this:


It handles the entire pipeline for you, producing a sequence of predicted tokens as output. You donâ€™t see the individual layers or components. This is great for fast, standard predictions, where you're only interested in the final result (e.g., generating text or answering a question).

### 2. **Manual Inspection (Step-by-Step Processing)**:
In the code you provided, you are accessing the model's components in a more granular way:

- **`model.model(input_ids)`** retrieves the output from the transformer layers **before** passing it through the final language modeling head (`lm_head`), which is used to map the hidden states to vocabulary logits.
  
- **`model.lm_head(model_output[0])`** then applies the final linear transformation (often called the "language modeling head") to generate the raw scores (logits) for each token in the vocabulary.

This approach allows you to inspect intermediate representations, such as the hidden states output by the transformer layers. This can be useful for:
- **Debugging**: Understanding which part of the model is behaving unexpectedly.
- **Fine-tuning/Custom Modifications**: You may want to modify the hidden states before they are passed to the `lm_head` for specific tasks.
- **Interpretability**: You can analyze the modelâ€™s internal representations to better understand how it is processing the input text, which might help in determining why certain predictions are made.

By doing this, you can also experiment with **alternative heads** (e.g., using a different layer instead of the `lm_head`) or even fine-tune specific components of the model.

In summary:
- **Direct inference** gives you fast and straightforward results.
- **Manual inspection** gives you access to how the model processes information internally, providing more control and understanding of its inner workings.

# Speeding up generation by caching keys and values


````python
prompt = "Write a very long email apologizing to Sarah for the tragic gardening mishap. Explain how it happened."

# Tokenize the input prompt
input_ids = tokenizer(prompt, return_tensors="pt").input_ids
input_ids = input_ids.to("cuda")
````

````python
%%timeit -n 1
# Generate the text
generation_output = model.generate(
  input_ids=input_ids,
  max_new_tokens=100,
  use_cache=True
)
````

````python
%%timeit -n 1
# Generate the text
generation_output = model.generate(
  input_ids=input_ids,
  max_new_tokens=100,
  use_cache=False
)
````
